```python
from kokoro import KPipeline
from IPython.display import display, Audio
import soundfile as sf
import torch


pipeline = KPipeline(lang_code='f') # <= make sure lang_code matches voice

# This text is for demonstration purposes only, unseen during training
text = '''
Aaaaah mon gars sûr, on va en faire des choses ensemble.

Aller maintenant je te raconte une histoire avec un extra terrestre fait de pâte à modeler.

Il était une fois un extra terrestre qui s'appelait Bob. Bob était un extra terrestre très spécial, car il était fait de pâte à modeler. Il était tout vert, avec des yeux globuleux et des antennes sur la tête. Il avait aussi des bras et des jambes, mais ils étaient très courts.

Bob vivait sur une planète très éloignée de la Terre. Sa planète s'appelait Zorg. Zorg était une planète très étrange, car elle était recouverte de pâte à modeler. Il y avait des montagnes de pâte à modeler, des rivières de pâte à modeler, et même des arbres de pâte à modeler.

Un jour, Bob décida de partir à l'aventure. Il prit son vaisseau spatial en forme de pâte à modeler et s'envola vers l'espace. Il voyagea pendant des jours et des jours, jusqu'à ce qu'il arrive sur la Terre.

Quand Bob atterrit sur la Terre, il fut très surpris. Il découvrit que les humains étaient très différents des extra terrestres de Zorg. Les humains étaient tout petits, avec des têtes rondes et des bras et des jambes très longs. Ils ne ressemblaient en rien à Bob.
'''
# text = '「もしおれがただ偶然、そしてこうしようというつもりでなくここに立っているのなら、ちょっとばかり絶望するところだな」と、そんなことが彼の頭に思い浮かんだ。'
# text = '中國人民不信邪也不怕邪，不惹事也不怕事，任何外國不要指望我們會拿自己的核心利益做交易，不要指望我們會吞下損害我國主權、安全、發展利益的苦果！'
# text = 'Los partidos políticos tradicionales compiten con los populismos y los movimientos asamblearios.'
# text = 'Le dromadaire resplendissant déambulait tranquillement dans les méandres en mastiquant de petites feuilles vernissées.'
# text = 'ट्रांसपोर्टरों की हड़ताल लगातार पांचवें दिन जारी, दिसंबर से इलेक्ट्रॉनिक टोल कलेक्शनल सिस्टम'
# text = "Allora cominciava l'insonnia, o un dormiveglia peggiore dell'insonnia, che talvolta assumeva i caratteri dell'incubo."
# text = 'Elabora relatórios de acompanhamento cronológico para as diferentes unidades do Departamento que propõem contratos.'

# 4️⃣ Generate, display, and save audio files in a loop.
generator = pipeline(
    text, voice='zf_xiaoxiao', # <= change voice here
    speed=1, split_pattern=r'\n+'
)

# Alternatively, load voice tensor directly:
# voice_tensor = torch.load('path/to/voice.pt', weights_only=True)
# generator = pipeline(
#     text, voice=voice_tensor,
#     speed=1, split_pattern=r'\n+'
# )

for i, (gs, ps, audio) in enumerate(generator):
    print(i)  # i => index
    print(gs) # gs => graphemes/text
    print(ps) # ps => phonemes
    display(Audio(data=audio, rate=24000, autoplay=i==0))
    sf.write(f'{i}.wav', audio, 24000) # save each audio file
```
