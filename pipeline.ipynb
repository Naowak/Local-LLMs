```python
from kokoro import KPipeline
from IPython.display import display, Audio
import soundfile as sf
import requests
import os
import json

# Set the environment variable within the notebook
os.environ['PHONEMIZER_ESPEAK_LIBRARY'] = '/opt/homebrew/opt/espeak-ng/lib/libespeak-ng.dylib'

# Make API request to get the LLM with streaming
response = requests.post(
    "http://localhost:1234/v1/chat/completions", 
    json={
        "messages": [{"role": "user", "content": "Tu peux me raconter ta journée ?"}],
        "prompt": "",
        "temperature": 0.9,
        "stream": True  # Enable streaming
    },
    stream=True  # Tell requests to stream the response
)

# Initialize pipeline for TTS
pipeline = KPipeline(lang_code='f')  # make sure lang_code matches voice

# Initialize buffer to accumulate text
buffer = ""
chunk_counter = 0

# Process streaming response
for chunk in response.iter_lines():
    if chunk:
        chunk_text = chunk.decode('utf-8')
        
        # Skip data: prefix if present
        if chunk_text.startswith("data: "):
            chunk_text = chunk_text[6:]
            
        if chunk_text.strip() == "[DONE]" or not chunk_text.strip():
            continue
            
        try:
            # Parse the JSON data
            data = json.loads(chunk_text)
            
            # Extract new text content
            new_text = ""
            if "choices" in data and len(data["choices"]) > 0:
                choice = data["choices"][0]
                if "delta" in choice and "content" in choice["delta"]:
                    new_text = choice["delta"]["content"]
                elif "text" in choice:
                    new_text = choice["text"]
                
            if new_text:
                # Add to buffer
                buffer += new_text
                
                # Check if we have a complete sentence
                if any(buffer.endswith(end) for end in ['.', '!', '?', '\n']):
                    print(f"Processing complete sentence: {buffer}")
                    
                    # Generate audio
                    generator = pipeline(
                        buffer, 
                        voice='ff_siwis',
                        speed=1
                    )
                    
                    for gs, ps, audio in generator:
                        print(gs)
                        print(ps)
                        display(Audio(data=audio, rate=24000, autoplay=True))
                        sf.write(f'chunk_{chunk_counter}.wav', audio, 24000)
                        chunk_counter += 1
                    
                    # Reset buffer
                    buffer = ""
        except json.JSONDecodeError:
            print(f"Could not parse JSON: {chunk_text}")

# Process any remaining text
if buffer:
    generator = pipeline(
        buffer, 
        voice='ff_siwis',
        speed=1
    )
    
    for gs, ps, audio in generator:
        print(f"Final chunk {chunk_counter}:")
        print(gs)
        print(ps)
        display(Audio(data=audio, rate=24000, autoplay=True))
        sf.write(f'chunk_{chunk_counter}.wav', audio, 24000)
        chunk_counter += 1
```
